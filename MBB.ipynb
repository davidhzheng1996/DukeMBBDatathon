```python
df_oppo = pd.read_excel("Opponent Data_2018.xlsx")
```

```python
df_oppo = df_oppo.assign(win = df_oppo.outcome == 'W')
```

```python
df1 = df_oppo[['opponent2','win']].groupby(['opponent2']).agg(['sum', 'count'])
```

```python
df1 = df1.assign(percentage = df1.win['sum']/df1.win['count'])
```

```python
df1 = df1.sort_values('percentage')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.scatter(df1.index, df1.percentage)
plt.xticks(df1.index, rotation='vertical')
plt.show()
```
